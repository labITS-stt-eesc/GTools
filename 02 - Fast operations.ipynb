# Features of GTools
GTools is a set of tools for complex urban network analysis. It mostly suplements OSMnx creating new sets of functions and making it work with iGraph to make some operations faster.

It also includes tools to add information from shapefiles on graph networks, this includes marking nodes with traffic zones and adding demographic data to nodes and edges.

Finally, it has some tools to make trip allocation from OD pairs easier.

# 2 - New formats of queries for OSM data.

Some funcitons that recieve a networkx graph (easy to use) and computes fast operations converting it to iGraph.

```python
import gtools as gt
import osmnx as ox
import networkx as nx
```

```python
G = gt.graph_from_circle('Curitiba, PR, Brasil', radius=2000, network_type='drive',) #circle centered in the address
ox.plot_graph(G)
```

### 1.1 - Fast centralities
Calculating centralities is generally very slow for large graphs, and iGraph is maybe an order of magnitude or more faster than NetworkG in that respect.

```python
# betweenness
print('NetworkX function:')
%timeit nx.edge_betweenness(G, weight='length')
print('\nFast operation')
%timeit gt.fast_betweenness(G, kind='edge', weight='length')
```

```python
#closeness
print('NetworkX function:')
%timeit nx.closeness_centrality(G, distance='length')
print('\nFast operation')
%timeit gt.fast_closeness(G, kind='node', weight='length')
```
